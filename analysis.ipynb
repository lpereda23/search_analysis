```python
import pandas 
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

Here I will be analyzing Google search trends on queries based on "Crypto" or any other keyword desired. 
Creating a datafram of the 10 top countries which search [Crypto]

```python
trends.build_payload(kw_list=["Crypto"])
data = trends.interest_by_region()
data = data.sort_values(by="Crypto", ascending=False)
data = data.head(10)
print(data)
```

According from the above results "Crypto" are mostly done in The Netherlands

```python
data.reset_index().plot(x="geoName", y="Crypto",
                       figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

Recently "Crypto" has been a really trendy and more and more people are investing in some sort of crypto. Let's have a look at the trend searches and see the how it has behaved. 

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Crypto'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20,15))
data["Crypto"].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Crypto', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
